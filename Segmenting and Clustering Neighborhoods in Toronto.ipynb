### Task 1. Getting all the dependencies.

```python
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files

#!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
!pip install geopy
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

#!conda install -c conda-forge folium --yes # uncomment this line if you haven't completed the Foursquare API lab
!pip install folium
import folium # map rendering library

print('Libraries imported.')
```

### Task 2. Downloading data through the link.

```python
table = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")

toronto_data = pd.DataFrame(table[0])
toronto_data.head()
```

### Task 3. Drop all rows with Not assigned Borough.

```python
toronto_data.drop(toronto_data[toronto_data['Borough']=='Not assigned'].index, inplace=True)
toronto_data.head()
```

#### Just reseting index.

```python
toronto_data.reset_index(inplace=True)
del toronto_data['index']
toronto_data
```

### Task 4. Replacing Non assigned neighborhood with borough.

```python
mask = toronto_data['Neighborhood'] == 'Non assigned'
toronto_data.loc[mask,'Neighborhood'] = toronto_data['Borough']
toronto_data.head()
```

### Task 5. Print number of rows.

```python
print('Number of rows is',toronto_data.shape[0])
```

### Task 6. Importing geo data from csv.

```python
geo_data = pd.read_csv("https://cocl.us/Geospatial_data")
geo_data.head()
```

#### Adding geo data to Toronto database.

```python
toronto_data = pd.merge(toronto_data, geo_data[['Postal Code','Latitude', 'Longitude']], left_on='Postal code', right_on='Postal Code', how='left')
toronto_data.head()
```

#### Removing duplicated column.

```python
del toronto_data['Postal Code']
toronto_data.head()
```

### Task 7. Creating map of neigborhoods of Toronto.

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[43.651070, -79.347015], zoom_start=11)

# add markers to map
for lat, lng, borough, neighborhood in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Borough'], toronto_data['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

### Task 8. Creating check if borough have 'Toronto' in its name.

```python
toronto_fit_data = toronto_data
toronto_fit_data['Have Toronto in name'] = toronto_fit_data['Borough'].str.contains(r'Toronto')
toronto_fit_data.head()
```

### Task 9. Fitting a model. I used 4 clusters for this purpose.

```python
# set number of clusters
kclusters = 4

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_fit_data[['Latitude','Longitude','Have Toronto in name']])

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

#### Adding lables to main table.

```python
toronto_data['Cluster']=kmeans.labels_
toronto_data.head()
```

#### Removing check columns.

```python
del toronto_data['Have \'Toronto\' in name']
del toronto_data['Have Toronto in name']
toronto_data.head()
```

### Task 10. Creating a map of Toronto neigborhoods, already clustered.

```python
# create map
map_clusters = folium.Map(location=[43.651070, -79.347015], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Neighborhood'], toronto_data['Cluster']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
