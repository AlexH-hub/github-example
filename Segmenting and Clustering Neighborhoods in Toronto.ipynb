### Task 2. Downloading data through the link.

```python
table = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")

toronto_data = pd.DataFrame(table[0])
toronto_data.head()
```

### Task 3. Drop all rows with Not assigned Borough.

```python
toronto_data.drop(toronto_data[toronto_data['Borough']=='Not assigned'].index, inplace=True)
toronto_data.head()
```

#### Just reseting index.

```python
toronto_data.reset_index(inplace=True)
del toronto_data['index']
toronto_data
```

### Task 4. Replacing Non assigned neighborhood with borough.

```python
mask = toronto_data['Neighborhood'] == 'Non assigned'
toronto_data.loc[mask,'Neighborhood'] = toronto_data['Borough']
toronto_data.head()
```
